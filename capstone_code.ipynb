## Visualising the data

```python
df['startdate'] = pd.to_datetime(df['startdate'])
df['month'] = df['startdate'].dt.month

plt.figure(figsize=(12, 6))
sns.boxplot(x='month', y='contest-tmp2m-14d__tmp2m', data=df)
plt.title('Box Plot of Mean Temperature by Month')
plt.xlabel('Month')
plt.ylabel('Mean Temperature (14 days)')
```

```python
# relationships between a selection of features using a pair plot.
features = [
    'lat', 
    'lon', 
    'contest-pevpr-sfc-gauss-14d__pevpr', 
    'nmme0-tmp2m-34w__cancm30', 
    'nmme0-tmp2m-34w__cfsv20', 
    'contest-tmp2m-14d__tmp2m'
]

missing_features = [feature for feature in features if feature not in df.columns]
if missing_features:
    print(f"Missing features: {missing_features}")
else:
    sns.pairplot(df[features])
    plt.suptitle('Pair Plot of Selected Features', y=1.02)
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='contest-pevpr-sfc-gauss-14d__pevpr', y='contest-tmp2m-14d__tmp2m', data=df)
plt.title('Scatter Plot of Potential Evaporation vs Mean Temperature')
plt.xlabel('Potential Evaporation (14 days)')
plt.ylabel('Mean Temperature (14 days)')
```

```python
# 1. Temperature Distribution and Model Comparison
plt.figure(figsize=(15, 6))

# Get mean temperature forecasts from different models
temp_models = [col for col in df.columns if 'tmp2m' in col and 'nmme0mean' in col]
temp_data = df[temp_models]

# Plot boxplot comparing different model predictions
plt.subplot(1, 2, 1)
temp_data.boxplot()
plt.xticks(rotation=45)
plt.title('Temperature Forecasts Across Models')
plt.ylabel('Temperature')

# Plot actual temperature distribution
plt.subplot(1, 2, 2)
sns.histplot(df['contest-tmp2m-14d__tmp2m'], kde=True)
plt.title('Actual Temperature Distribution')
plt.xlabel('Temperature')

plt.tight_layout()
plt.show()
```

```python
# 2. Geographic Distribution of Measurements with Temperature
plt.figure(figsize=(15, 6))

# Create scatter plot of locations colored by temperature
plt.subplot(1, 2, 1)
plt.scatter(df['lon'], df['lat'], 
           c=df['contest-tmp2m-14d__tmp2m'], 
           cmap='coolwarm',
           alpha=0.5)
plt.colorbar(label='Temperature')
plt.title('Geographic Distribution of Temperature')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add density plot
plt.subplot(1, 2, 2)
plt.hist2d(df['lon'], df['lat'], bins=50, cmap='viridis')
plt.colorbar(label='Measurement Density')
plt.title('Density of Measurements')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.tight_layout()
plt.show()
```

```python
# 3. Temporal Patterns and Seasonality
plt.figure(figsize=(15, 6))

# Monthly temperature patterns
plt.subplot(1, 2, 1)
sns.boxplot(x='month', y='contest-tmp2m-14d__tmp2m', data=df)
plt.title('Monthly Temperature Distribution')
plt.xlabel('Month')
plt.ylabel('Temperature')

# Time series of temperatures (sample of locations)
plt.subplot(1, 2, 2)
sample_locations = df.groupby(['lat', 'lon']).first().reset_index().head(5)
for _, loc in sample_locations.iterrows():
    loc_data = df[(df['lat'] == loc['lat']) & (df['lon'] == loc['lon'])]
    plt.plot(loc_data['startdate'], loc_data['contest-tmp2m-14d__tmp2m'], 
             label=f'({loc["lat"]:.1f}, {loc["lon"]:.1f})', alpha=0.7)
plt.title('Temperature Time Series for Sample Locations')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# 5. Temperature-Precipitation Relationships and Environmental Factors
fig = plt.figure(figsize=(15, 5))

# Temperature vs Precipitation
plt.subplot(1, 3, 1)
sns.scatterplot(data=df, 
                x='nmme0-tmp2m-34w__nmme0mean',
                y='nmme0-prate-34w__nmme0mean',
                alpha=0.5)
plt.title('Temperature vs Precipitation')
plt.xlabel('Temperature Forecast')
plt.ylabel('Precipitation Forecast')

# Temperature by Latitude
plt.subplot(1, 3, 2)
sns.scatterplot(data=df,
                x='lat',
                y='contest-tmp2m-14d__tmp2m',
                alpha=0.5)
plt.title('Temperature by Latitude')
plt.xlabel('Latitude')
plt.ylabel('Temperature')

# Add trend line for temperature vs precipitation
plt.subplot(1, 3, 3)
sns.regplot(data=df,
            x='nmme0-tmp2m-34w__nmme0mean',
            y='nmme0-prate-34w__nmme0mean',
            scatter_kws={'alpha':0.5})
plt.title('Temperature-Precipitation Trend')
plt.xlabel('Temperature Forecast')
plt.ylabel('Precipitation Forecast')

plt.tight_layout()
plt.show()
```
